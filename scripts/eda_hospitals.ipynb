```python
import os

candidates = ["original-data/baptist.csv", "../original-data/baptist.csv"]
for p in candidates:
    if os.path.exists(p):
        df_baptist = pd.read_csv(p, skiprows=2)
        print(f"Loaded {p} ({df_baptist.shape[0]} rows, {df_baptist.shape[1]} cols)")
        break
else:
    raise FileNotFoundError("baptist.csv not found in any of: " + ", ".join(candidates))
```

```python
import os

input_file = "../original-data/baptist.csv"
output_file = "baptist_top100.csv"

# Automatically create the 'samplefiles' folder if it doesn't exist
output_dir = os.path.dirname(output_file)
if output_dir and not os.path.exists(output_dir):
    os.makedirs(output_dir)

with open(input_file, "r", encoding="utf-8", errors="ignore") as f_in:
    with open(output_file, "w", encoding="utf-8") as f_out:
        for i in range(100):
            line = f_in.readline()
            if not line:  # Stop if the file has fewer than 100 lines
                break
            f_out.write(line)

print(f"Successfully exported the top 100 rows to {output_file}")
```
